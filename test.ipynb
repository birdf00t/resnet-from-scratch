```python
import torch
import torch.nn.functional as F

class Conv2d:
    def __init__(self, in_channels, out_channels, kernel_size):
        self.kernel_size = kernel_size
        self.out_channels = out_channels
        self.weight = torch.randn(out_channels, in_channels, kernel_size, kernel_size)
        

    def forward(self,x, padding=0,stride =1):
        x = self.pad2d(x,padding=padding)
        height_out = int((x.shape[2] - self.kernel_size)/stride + 1)
        weight_out = int((x.shape[3] - self.kernel_size)/stride + 1)
        out_x = torch.zeros(x.shape[0],self.out_channels, height_out,weight_out )


    def pad2d(self,x,padding):
        x= F.pad(x, (padding, padding, padding, padding))
        return x
```
